```python
#import dependencies
import pandas as pd
```

```python
# load files
files_to_load = "Resources/missing_grades.csv"
```

```python
# read the csv into a DF
missing_grade_df = pd.read_csv(files_to_load)
missing_grade_df
```

```python
# check the mean with missing values not changed
# note if the column has no space, you can directly call it like I do for reading score below.
print("Reading Score Mean: " + str(missing_grade_df.reading_score.mean().round(2)))
print("Math Score Mean: " + str(missing_grade_df['math_score'].mean().round(2)))
```

```python
# cleaning method #1 - drop NaN
# rows with any column NaN are dropped
missing_grade_dfdrop = pd.read_csv(files_to_load)
drop = missing_grade_dfdrop.dropna()
drop
```

```python
# check the mean with missing values not changed
print("Reading Score Mean: " + str(drop['reading_score'].mean().round(2)))
print("Math Score Mean: " + str(drop['math_score'].mean().round(2)))
```

```python
# cleaning method #2 - fill in NaN with values
# fill empty rows with reading or math score of 85
missing_grade_dffill = pd.read_csv(files_to_load)
fill = missing_grade_dffill.fillna(85)
fill
```

```python
# check new means for fillna
# check the mean with missing values not changed
print("Reading Score Mean: " + str(fill['reading_score'].mean().round(2)))
print("Math Score Mean: " + str(fill['math_score'].mean().round(2)))
```
